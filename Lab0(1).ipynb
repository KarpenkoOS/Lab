```python
import numpy as np
import pprint
np.set_printoptions(precision=5)
pp = pprint.PrettyPrinter(indent=4)
```

```python
P = np.array([[0, 0.3, 0.7, 0],
[0.5, 0, 0.5, 0],
[0, 0.5, 0, 0.5],
[0, 0, 0.5, 0.5]])
```

```python
for i in range(501):
    pp.pprint(np.linalg.matrix_power(P, i))
    print("."*50)
```
